```python
class Set:
    
    def __init__(self,l=list()):
        self.s = list()
            
        for i in l:
            if i not in self.s:
                self.s.append(i)
        
    
    def add(self,elem): #set끼리의 합이 아님
        
        elem = str(elem)
        
        if elem not in self.s:
            self.s.append(elem)
    
    def discard(self,elem):
        
        elem = str(elem)
        
        if elem in self.s:
            self.s.remove(elem)
            
     
    def clear(self):
        
        self.s = []
        
    def __len__(self):
        
        i = 0
        count = 0
        
        for i in self.s:
            
            count += 1
        
        return count
    
    def __str__(self):
        
        
        for i in range(len(self.s)):
            self.s[i] = str(self.s[i])
            
        ss = ",".join(self.s)
        sss = '{'+ ss + '}'
        return sss
    

    def __contains__(self,elem):
        
        elem = str(elem)
        
        if elem in self.s:
            return True
        
        return False
    
    def __le__(self,other):
        
        le = list()
        for i in self.s:
            if i in other.s:
                le.append(i)
                if len(le) == len(self.s):
                    return True
        
        return False
    
    def __ge__(self,other):
        
        ge = list()
        for i in other.s:
            if i in self.s:
                ge.append(i)
                if len(ge) == len(other.s):
                    return True 

        return False
    
    def __or__(self,other):
        
        tmp_list = list()
        
        for i in self.s:
            tmp_list.append(i)
            
        for i in other.s:
            if i not in tmp_list : tmp_list.append(i)
                
        return Set(tmp_list)
        #return Set(self.s + ohters.s)
        
    def __and__(self,other):
        
        and_list = list()
        
        for i in self.s:
            if i in other.s:
                and_list.append(i)
        
        return Set(and_list)
    
    def __sub__(self,other):
        
        and_list = list()
        sub_list = self.s + list()
        
        
        for i in self.s:
            if i in other.s:
                and_list.append(i)
                
        for j in and_list:
            if j in self.s:
                self.s.remove(j)
                
        return self
        
    def __ior__(self,other):
        tmp_list = list()
        self_ss = self.s
        
        for i in self.s:
            tmp_list.append(i)
            
        for i in other.s:
            if i not in tmp_list : tmp_list.append(i)
         
        self_cp = self.s.copy()
        return Set(tmp_list)
    
    def __iand__(self,other):
        
        and_list = list()
        
        for i in self.s:
            if i in other.s:
                and_list.append(i)
        
        self.s = and_list;

        return self
    
    
    
    def __isub__(self,other):
        
        and_list = list()
        
        print("this 주소~~", id(self))
        
        for i in self.s:
            if i in other.s:
                and_list.append(i)
                
        for j in and_list:
            if j in self.s:
                self.s.remove(j)

        return self



```
